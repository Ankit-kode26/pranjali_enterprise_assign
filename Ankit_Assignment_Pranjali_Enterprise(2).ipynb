# 2.) Steps for preparing Option data:

```python
import pandas as pd 
#Reading the data into Data frame 
read_data = pd.read_csv("C:/Users/91981/OneDrive/Desktop/GFDLNFO_07042017.csv")
read_data.head(10)
```

```python
read_data.info()
```

### 2.1) Filter BANKNIFTYxxxxxxxxxxCE.NFO & BANKNIFTYxxxxxxxxxxPE.NFO from ticker column.

```python
pattern = r'BANKNIFTY\d{2}[A-Z]{3}\d{7}(CE|PE)'
read_data = read_data[read_data['Ticker'].str.match(pattern)]
read_data.head(20)
```

```python
read_data.reset_index(drop=True, inplace=True)
read_data.head(10)
```

```python
read_data.tail(10)
```

### 2.1.1) Extracting scrip from the ticker

```python
read_data['scrip1'] = read_data['Ticker'].str.extract(r'(\d+PE|\d+CE)')
read_data['scrip'] = read_data['scrip1'].str[2:]
read_data['scrip'].head(10)
read_data = read_data.drop(columns= ['scrip1'])
read_data.head(10)
```

### 2.1.2) Extracting Expiry date from the ticker 

```python
read_data['expiry'] = read_data['Ticker'].str.extract(r'(\d+[A-Z]+\d{2})')
read_data['expiry'].head(10)
read_data.head(10)
```

```python
# Converting expiry column into Datetime format
read_data['expiry'] = pd.to_datetime(read_data['expiry'])
read_data['expiry'].info()
```

```python
read_data.head(5)
```

### 2.2) Convert Date and Time into a new datetime column


```python
read_data['datetime'] = pd.to_datetime(read_data['Date']+ ' ' + read_data['Time'])
print(read_data['datetime'].head(10))
```

```python
read_data.head(5)
```

### 2.3) Calculate dte = expiry - Date

```python
read_data['Date'].info()
```

```python
read_data['Date'] = pd.to_datetime(read_data['Date'])
read_data['Date'].info()
```

```python
read_data['expiry'].info()
```

```python
read_data['dte'] = read_data['Date'] - read_data['expiry']
read_data['dte'].head(10)
```

```python
read_data.head(10)
```

### 2.5) Remove columns: Ticker, Date, Time, Volume, Open Interest.


```python
remove_columns = ['Ticker','Date','Time','Volume','Open Interest']
read_data = read_data.drop(columns= remove_columns)
```

```python
read_data.head(10)
```

### 2.6) Sort data by datetime.

```python
read_data_sorted = read_data.sort_values(by='datetime', ascending=True)
```
