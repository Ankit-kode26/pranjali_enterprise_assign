# 1.) Steps for preparing Future data:


```python
import pandas as pd 
#Reading the data into Data frame 
read_data = pd.read_csv("C:/Users/91981/OneDrive/Desktop/GFDLNFO_07042017.csv")
read_data.head(10)
```

```python
read_data.info()
```

### 1.1) Filter BANKNIFTY-I.NFO from Ticker column.

```python
# Here we are filtering the Dataframe read_data
# Excluding the rows from the 'Ticker' column that contain the value 'BANKNIFTY-I.NFO' , and assigning the dataframe back to the read_data
read_data=read_data[read_data['Ticker']!='BANKNIFTY-I.NFO'] 
```

```python
read_data['Ticker'].head(10)
```

### 1.2) Convert Date and Time into a new datetime column

```python
print(read_data['Date'].head(10))
# There DataType here in Date Column is Object we have to change it to DateTime 
```

```python
print(read_data['Time'].head(10))
# There DataType here in Time Column is Object we have to change it to DateTime 
```

```python
# Here We will concatenate Date and Time columns and change their datatype into DateTime
# We will change the foemayt of the datetime column by using strftime function and define the required format 
read_data['datetime'] = pd.to_datetime(read_data['Date']+ ' ' + read_data['Time'])
print(read_data['datetime'].head(10))
```

```python
read_data.head(10)
```

### 1.3) Remove columns: Ticker, Date, Time, Volume, Open Interest.

```python
# Here we will add columns in list name 'remove_columns' and then we will use DROP FUNCTION to remove those columns from our Data 
remove_columns = ['Ticker','Date','Time','Volume','Open Interest']
read_data = read_data.drop(columns= remove_columns)
```

```python
# Expected columns are removed from the Data
read_data.head(10)
```

### 1.4) Sort data by datetime.

```python
read_data_sorted = read_data.sort_values(by='datetime')
read_data_sorted.head(5)
```
